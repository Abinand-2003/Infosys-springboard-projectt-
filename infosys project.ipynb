```python
import requests
import os

API_KEY = "API KEY"
"""
 I hardcoded my api key and when i run the code it shows this error Recommendations:
Unexpected response format: {'error': {'message': 'More credits are required to run this request. 16384 token capcity required, 4000 available. To increase, visit https://openrouter.ai/credits and upgrade to a paid account', 'code': 402}}

"""
BASE_URL = "https://openrouter.ai/api/v1/chat/completions"

def get_hospitality_recommendations(customer_profile):
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json",
    }
    payload = {
        "model": "openai/gpt-4o-2024-11-20",
        "messages": [
            {"role": "system", "content": "You are an expert in hospitality and guest experience personalization of a particular hotel. Recommend suitable amenities and dining options based on customer profiles to customers who are coming to a partiular hotel."},
            {"role": "user", "content": f"""
                Customer Profile:
                {customer_profile}
                
                Based on this profile, recommend hotel amenities such as sauna, bar, swimming pool, and dining styles that align with their preferences.
            """}
        ]
    }
    try:
        response = requests.post(BASE_URL, headers=headers, json=payload)
        response.raise_for_status()
        data = response.json()
        if "choices" in data:
            return data["choices"][0]["message"]["content"]
        else:
            return f"Unexpected response format: {data}"
    except requests.exceptions.RequestException as e:
        return f"An error occurred: {e}"

if __name__ == "__main__":
    customer_profile = """
    - Age: 40
    - Traveling for leisure
    - Prefers luxury experiences
    - Non-vegetarian
    - Enjoys spa and wellness activities
    - Interested in nightlife and fine dining
    - Traveling as a couple
    """
    recommendations = get_hospitality_recommendations(customer_profile)
    print("Recommendations:")
    print(recommendations)


```
